```python
import re 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
f = open('brothers.txt' , encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r"\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s\w+\s-\s"
```

```python
messages = re.split(pattern , data)[1:]
len(messages)
print(messages)
```

```python
dates = re.findall(pattern , data)
len(dates)
print(dates)
```

```python
times_ = [] 
dates_  = []
for i in dates:
    dates_.append(i.split(", ")[0])
    times_.append(i.split(", ")[1])

print(dates_)
print(times_)
```

```python
time_ = []

for i in times_:
    time_.append(i.split("\u202f")[0])
```

```python
print(time_)
```

```python
df = pd.DataFrame({
    "user_message" : messages,
    "date" : dates_,
    "time" : time_
})
```

```python
df
```

```python
# splitting user name and message 

user = []
msg = []

for i in df["user_message"]:
    x = re.split(r"([\w\W]+?):\s", i) # splitting based on format 
    if x[1:]: # if exists 
        user.append(x[1])
        msg.append(x[2])
    else: # if not exists 
        user.append('Group notification')
        msg.append(x[0])
```

```python
df['user'] = user
df['msg'] = msg 
df.drop(columns = ['user_message'] , inplace = True)
```

```python
# Converting to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extracting  year, month name, and day name
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()  # Corrected: directly using month_name() on the datetime series
df['day'] = df['date'].dt.day_name()


print(df[['date', 'year', 'month', 'day']].head())
```

```python
selected_user = input("enter the user name out of {}".format(df['user'].unique().tolist()))
```

```python
from urlextract import URLExtract
extract = URLExtract()
```

```python
def fetch_stats(selected_user , df ):

    if selected_user != "OverAll":
        df = df[df['user'] == selected_user]

    #number of messages
    num_msgs = df.shape[0]

    #number of words
    words = []
    for msg in df['user']:
        words.extend(msg.split(' '))

    #number of media
    num_med = df[df['msg'] == '<Media omitted>\n'].shape[0]

    #number of links
    links = []
    for msg in df['msg']:
        links.extend(extract.find_urls(msg))  # library used above to extract the links

    return num_msgs , len(words) , num_med , len(links)
```

```python
num_msgs , words , num_med , links =fetch_stats(selected_user , df)
```

```python
def monthly_timeline(selected_user , df):

    if selected_user != 'OverAll':
        df = df[df['user'] == selected_user]

    timeline = df.groupby(["year" , 'month'])['msg'].count().reset_index()

    time = []
    for i in range(timeline.shape[0]):
        time.append(timeline['month'][i] + '-' + str(timeline['year'][i]))

    timeline['time'] = time

    return timeline
```

```python
# monthly timeline plot
timeline = monthly_timeline(selected_user , df)

fig , ax = plt.subplots()
ax.plot(timeline['time'] , timeline['msg'] , color = 'maroon')
plt.xticks(rotation = 90)
plt.show()
```

```python
def daily_timeline(selected_user , df):
    if selected_user != 'OverAll':
        df = df[df['user'] == selected_user]

    timeline = df.groupby(df['date'])['msg'].count().reset_index()

    return timeline
```

```python
#daily timeline plot
timeline = daily_timeline(selected_user , df )
fig , ax = plt.subplots()
ax.plot(timeline['date'] , timeline['msg'] , color = 'purple')
plt.xticks(rotation = 90)
plt.show()
```

```python
def activity_map(selected_user , df ):
    if selected_user != 'OverAll':
        df = df[df['user'] == selected_user]

    activity_month_df = df.groupby('month')['msg'].count().reset_index()
    month_list = activity_month_df['month'].tolist()
    monthly_msg_list = activity_month_df['msg'].tolist()

    active_day_df = df.groupby('day')['msg'].count().reset_index()
    day_list = active_day_df['day'].tolist()
    day_msg_list = active_day_df['msg'].tolist()

    return activity_month_df , month_list , monthly_msg_list , active_day_df , day_list , day_msg_list
```

```python
activity_month_df , month_list  , monthly_msg_list , active_day_df , day_list , day_msg_list = activity_map(selected_user , df)
```

```python
fig , ax = plt.subplots()
ax.bar(activity_month_df['month'] , activity_month_df['msg'])
ax.bar(month_list[monthly_msg_list.index(max(monthly_msg_list))] , max(monthly_msg_list) ,
        color = 'green' , label = 'highest')
ax.bar(month_list[monthly_msg_list.index(min(monthly_msg_list))], min(monthly_msg_list), color='red',
        label = "lowest")
plt.xticks(rotation = 90)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(active_day_df['day'], active_day_df['msg'])
ax.bar(day_list[day_msg_list.index(max(day_msg_list))], max(day_msg_list), color='green',
       label='highest')
ax.bar(day_list[day_msg_list.index(min(day_msg_list))], min(day_msg_list), color='red',
       label="lowest")
plt.xticks(rotation=90)
plt.show()
```

```python
def most_common_words(selected_user , df):

    f = open('stop_hinglish.txt' , 'r')
    stop_words = f.read()

    if selected_user != 'OverAll':
        df = df[df['user'] == selected_user]

    temp = df[df['user'] != 'Group notification']
    temp = temp[temp['msg'] != '<Media omitted>']

    words = []

    for message in temp['msg']:
        for word in message.lower().split():
            if word not in stop_words:
                words.append(word)

    most_common_words_df = pd.DataFrame(Counter(words).most_common(20))
    return most_common_words_df
```

```python
most_common_words_df = most_common_words(selected_user , df)
```

```python
# most common words
plt.figure(figsize = (10 , 6))
plt.barh(most_common_words_df[0] , most_common_words_df[1])
plt.show()
```

```python
def most_common_emojis(selected_user, df):
    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]
        
    emojis = []
    for message in df['msg']:
        emojis.extend([c for c in message if c in emoji.EMOJI_DATA])  # Collect emojis from each message
    
    return emojis
```

```python
def activity_heatmap(selected_user , df):

    if selected_user != "OverAll":
        df = df[df['user'] == selected_user]

    daily_activity = df.groupby(df['date'].dt.date).size().reset_index(name='message_count')
    # Set 'date' column as index and fill missing dates with zero messages
    daily_activity = daily_activity.set_index('date').asfreq('D', fill_value=0)
    daily_activity['year'] = daily_activity.index.year
    daily_activity['month'] = daily_activity.index.month
    heatmap_data_ = daily_activity.pivot_table(values='message_count', index='month', columns='year', aggfunc='sum')

    return heatmap_data_
```

```python
heatmap_data = activity_heatmap(selected_user, df)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))  # update figure creation here
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title('Message Activity Heatmap')
plt.xlabel('Year')
plt.ylabel('Month')
plt.yticks(ticks=range(1, 13),
           labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
